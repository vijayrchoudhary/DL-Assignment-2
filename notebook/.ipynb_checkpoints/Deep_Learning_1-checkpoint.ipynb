```python
## Code to train the model ##

# Reading images from the folder
data_path='/content/Latest/'
categories=os.listdir(data_path)
labels=[i for i in range(len(categories))]
label_dict= dict(zip(categories,labels)) 

data=[]
target=[]

# For each class to be classified 
for category in categories:
    folder_path=os.path.join(data_path,category)
    img_names=os.listdir(folder_path)

    # For each image that belongs to current class    
    for img_name in img_names:
        img_path=os.path.join(folder_path,img_name)
        # Reading image in 224 X 224 dimension to match mobilenet input requirement
        image=load_img(img_path, target_size=(224, 224)) 
        # image is converted to array and preprocessed same as it did in imagenet training
        image=img_to_array(image)
        img=preprocess_input(image)
        try:
            # Adding training data and label to array 
            data.append(img)            
            target.append(label_dict[category])
            
        except Exception as e:
            print('Exception:',e)
            #if any exception rasied, the exception will be printed here. And pass to the next image

# data is normalized by max value in RGB representation
data=np.array(data)/255.0
target=np.array(target)
# converting target label to one hot vector
new_target=np_utils.to_categorical(target)

# building model as defined earlier
#model = build_model(224)

# Test data will be 20% of overall data
train_data,test_data,train_target,test_target=train_test_split(data,new_target,test_size=0.2,stratify=new_target)

# Validation data is 20% of training data
train_data_val,test_data_val,train_target_val,test_target_val=train_test_split(train_data,train_target,test_size=0.2,stratify=train_target)

```

#**PART 2**

```python
# Code to build and manipulate mobile net model

def build_model(target_size):
    input_tensor = Input(shape=(target_size, target_size, 3))
    # get mobilenet model from keras without top layer
    base_model = MobileNetV2(
        include_top=False,
        weights='imagenet',
        input_tensor=input_tensor)

    # trainable has to be false in order to freeze the layers
    for layer in base_model.layers:
        layer.trainable = False  

    # Add additional layers on top of model
    headModel = base_model.output
    headModel = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(headModel)    
    headModel = Conv2D(100,(3,3))(headModel)
    # flattening layers to fetch into dense layer
    headModel = Flatten(name="flatten")(headModel)
    # Dense layer of 128 neurons and relu activation function
    headModel = Dense(32, activation="relu")(headModel)
    # dropping inputs to avoid over fitting
    headModel = Dropout(0.1)(headModel)
    headModel = Dense(10, activation="relu")(headModel)
    # headModel = GlobalMaxPooling1D(name="scalar_output")(headModel)    
    #Softmax layer for final classification
    headModel = Dense(2, activation="softmax")(headModel)

    output_tensor = headModel
    model = Model(inputs=base_model.input, outputs=output_tensor)

    return model
```

```python
def build_analysis_model(target_size):
    input_tensor = Input(shape=(target_size, target_size, 3))
    # get mobilenet model from keras without top layer
    base_model = MobileNetV2(
        include_top=False,
        weights='imagenet',
        input_tensor=input_tensor)

    # trainable has to be false in order to freeze the layers
    for layer in base_model.layers:
        layer.trainable = False  

    # Add additional layers on top of model
    headModel = base_model.output
    headModel = GlobalMaxPooling2D(name="scalar_output")(headModel)    
    #Softmax layer for final classification
    headModel = Dense(2, activation="softmax")(headModel)

    output_tensor = headModel
    model = Model(inputs=base_model.input, outputs=output_tensor)

    return model
```

```python
from tensorflow.keras import backend as k
model = build_analysis_model(224)
print(model.summary())


```

```python

# model = models.load_model("/content/model.h5")
last_layer = model.get_layer("scalar_output")
layer_output = last_layer.output
layer_input = model.input
output_func = k.function([layer_input], layer_output)

classA = np.where(test_target_val[:,0] == 1)
classB = np.where(test_target_val[:,1] == 1)

classA_scalarOutput = output_func([test_data_val[classA]])
classB_scalarOutput = output_func([test_data_val[classB]])

# print(layerOutput)
# tmp_model = Model(model.layers[0].input, model.layers[i].output)
# tmp_output = tmp_model.predict(img)[0]
# layer_outputs.append(tmp_output)
```

```python
classA_neurons_allInput = np.where(classA_scalarOutput>4)   # indexes where scalar output is more than 4                  
classA_neurons_freq = np.bincount(classA_neurons_allInput[1])              # count frequencies of which neurons produced output max number of time
classA_neurons = np.where(classA_neurons_freq > 0.99 * len(classA[0]))
# print(classA_neurons)

classB_neurons_allInput = np.where(classB_scalarOutput>4)   # indexes where scalar output is more than 4                  
classB_neurons_freq = np.bincount(classB_neurons_allInput[1])              # count frequencies of which neurons produced output max number of time
classB_neurons = np.where(classB_neurons_freq > 0.99 * len(classB[0]))
# print(classB_neurons)

classA_identifier = np.setdiff1d(classA_neurons,classB_neurons)
classB_identifier = np.setdiff1d(classB_neurons,classA_neurons)

print(classA_identifier)
print(classB_identifier)
```

```python
relu_layer = model.get_layer(name="out_relu")
output_func = k.function([model.input],relu_layer.output)
output_A = output_func([test_data_val[classA[0]]])
print(output_A.shape)
output_B = output_func([test_data_val[classB[0]]])
print(output_B.shape)
```

```python
images_per_row = len(classA_identifier) 
n_features=len(classA_identifier) # Number of features in the feature map
size = output_A.shape[1] #The feature map has shape (1, size, size, n_features).
n_cols = n_features // images_per_row # Tiles the activation channels in this matrix

for img_ind in range(0,len(classA[0])):
    display_grid = np.zeros((size * n_cols, images_per_row * size))
    for col in range(n_cols): # Tiles each filter into a big horizontal grid
        for row in range(images_per_row):
            channel_image = output_A[img_ind, :, :, col * images_per_row + row]            
            display_grid[col * size : (col + 1) * size, # Displays the grid
                          row * size : (row + 1) * size] = channel_image

    scale = 1. / size

    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    # plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
```
